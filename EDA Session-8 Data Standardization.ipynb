# Standardization

- Standardization means scaling the data into one scale
- we have different columns has different units so that the value will vary
- One column has very huge values
- Another column has very less values
- So it is important to scale all types of units under one scale
- We have 2 procedures
- Standardization
    - Z-score:
        
        Z = (x-u) / sigma
        
        - The values ranges from -3 to 3
        
        
- Normalization
    - Min Max Scalar
    
    normalized = (x - xmin)/(x max - xmin)
    
    - The values ranges from 0 to 1

```python
# step -1  Take the prevailing wage column
#  Z-score

# 2.  mean of prevailing wage
# 3. standard devaiation of prewailig wage
# 4  Nr = pwage - mean
# 5  zscore = Nr/dr

```

```python
pwage = visa_df["prevailing_wage"]
pwage_mean = visa_df["prevailing_wage"].mean()
pwage_sd = visa_df["prevailing_wage"].std()
Nr = pwage - pwage_mean
visa_df["prevailing_wage_z"] = Nr/pwage_sd
```

**Normalization**
- min max scalar

```python
##  Read the data again

path = "C:\\Users\\chand\\onedrive\\Documents\\datascience\\Python Data files for Analysis\\Visadataset.csv"
visa_df = pd.read_csv(path)
visa_df.head()
```

```python
##  x-xmin /(x_max-m_min)

##   step-1 Read the pwage column
##   step-2 Find the min value
##   step-3 Find the max value
##   step-4 nr = x - x_min
##   step-5 dr = x_max - x_min
##   step_6  nr/dr




pwage = visa_df["prevailing_wage"]
pmin = visa_df["prevailing_wage"].min()
pmax = visa_df["prevailing_wage"].max()
nr = pwage - pmin
dr = pmax - pmin
visa_df["prevailing_wage_mm"] = nr/dr
```

```python
visa_df["prevailing_wage_mm"].max(), visa_df["prevailing_wage_mm"].min()
```
