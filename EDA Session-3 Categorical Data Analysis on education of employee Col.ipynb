```python
uni_lab = visa_df["education_of_employee"].unique()
count = []
for i in uni_lab:
    emp = visa_df["education_of_employee"] == i
    count.append(len(visa_df[emp]))
    
employee_df = pd.DataFrame(zip(uni_lab,count),columns = ["education_of_employee","count"])
employee_df
    
```

```python
employee_vc = visa_df["education_of_employee"].value_counts()
l1 = employee_vc.keys()
l2 = employee_vc.values
employee_vc_df = pd.DataFrame(zip(l1,l2),columns = ["education_of_employee","count"])
employee_vc_df
```

# Bar chart

```python
employee_vc_df
```

```python
plt.bar("education_of_employee","count",data = employee_vc_df)
plt.xlabel("education_of_employee")
plt.ylabel("count")
plt.title("edu_of_emp bar chart")
plt.savefig("education_of_emp_bar.jpg")
plt.show()
```

```python
l = visa_df["education_of_employee"].value_counts().keys()
sns.countplot(data = visa_df,
             x = "education_of_employee",
             order = l)
plt.xlabel("education_of_employee")
plt.ylabel("count")
plt.title("edu_of_emp bar chart using count_values")
plt.savefig("education_of_emp_bar1.jpg")
plt.show()
```
