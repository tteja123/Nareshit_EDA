## Outlier Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "C:\\Users\\chand\\Onedrive\\Documents\\datascience\\Hackathon\\loan_prediction.csv"
loan_df = pd.read_csv(path)
loan_df.head()
```

```python
##==================   Step - 1 ===========================

Q1 = np.percentile(loan_df["ApplicantIncome"],25)
Q2 = np.percentile(loan_df["ApplicantIncome"],50)
Q3 = np.percentile(loan_df["ApplicantIncome"],75)

##========================  Step - 2 =============================

IQR = Q3 - Q1

##======================== Step - 3 ==============================

lb = Q1 - 1.5 * IQR
ub = Q3 + 1.5 * IQR

##=======================  Step - 4 ==============================

c1 = loan_df["ApplicantIncome"] < lb
c2 = loan_df["ApplicantIncome"] > ub
con = c1 | c2
outliers_df = loan_df[con]
#print(outliers_df)

##======================  Step - 5 =============================

c1 = loan_df["ApplicantIncome"] > lb
c2 = loan_df["ApplicantIncome"] < ub
con = c1 & c2
non_outliers_df = loan_df[con]
non_outliers_df
```

```python
plt.hist(loan_df["ApplicantIncome"])
plt.hist(non_outliers_df["ApplicantIncome"])
plt.show()
```

```python
plt.boxplot(loan_df["ApplicantIncome"])
plt.boxplot(non_outliers_df["ApplicantIncome"])
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.subplot(2,2,1)
plt.hist(loan_df["ApplicantIncome"],bins = 40)
plt.subplot(2,2,2)
plt.boxplot(loan_df["ApplicantIncome"])
plt.subplot(2,2,3)
plt.hist(non_outliers_df["ApplicantIncome"], bins = 40)
plt.subplot(2,2,4)
plt.boxplot(non_outliers_df["ApplicantIncome"])
plt.show()
```

```python
##==================   Step - 1 ===========================

Q1 = np.percentile(loan_df["ApplicantIncome"],25)
Q2 = np.percentile(loan_df["ApplicantIncome"],50)
Q3 = np.percentile(loan_df["ApplicantIncome"],75)

##========================  Step - 2 =============================

IQR = Q3 - Q1

##======================== Step - 3 ==============================

lb = Q1 - 1.5 * IQR
ub = Q3 + 1.5 * IQR

median = loan_df["ApplicantIncome"].median()

list1 = []
for i in loan_df["ApplicantIncome"]:
    if i < lb or i > ub:
        list1.append(median)
    else:
        list1.append(i)
        
loan_df["ApplicantIncome_new"] = list1
```
